### Problem One: Rock Paper Scissors

Description: A Rock Paper Scissors game with an adaptive computer depending upon the User's Choice

```python
#Part A
import re
#function to add the two lists together
def addlists(quote,author):
    list = [str(quote[i])+ " -" + str(author[i]) for i in range(len(author))]
    return list

def listFullOfQuotes():
    list1 = []
    f = open('quotes.txt','r')
    for line in f:
        #replace weird ` character with apostrophe 
        line = line.replace("`", "'")
        list1.append(re.split("[\n]",line)[0])
    f.close()
    return addlists(list1[0::2], list1[1::2])
def main():
    out = listFullOfQuotes()  
    print out
main()
```

#### b.
    -  wordsFromFullQuotes:
        Takes a formatted string and turns it into words. Handles for weird exception in dataset
        Input:
            Formatted Quote-Author string
        Ouput:
            A cleaned list of words in quotes

```python
def wordsFromFullQuotes(string):
    item = string.lower()
    out = [item.split("-").pop()]
    for i in out:
        s = '-' + out[0]
        item = item.replace(s,"")
        k = item.split('[\-=~!@#$%^&*()+\[\]{};\\:"|<,./<>?|\s]')
        for index, item in enumerate(k):
            k[index] = item.rstrip("'")
            k[index] = k[index].lstrip("'")
    outter= k + out

    while "" in outter:-
        outter.remove("")
    #clear out random one letter words
    for item in outter:
        if len(item) == 1 and item.lower() not in ['a','i','o']:
            outter.remove(item)
    return outter
def main():
    print wordsFromFullQuotes("hi, my name is is Gaby da 901928_ -Gaby Lio")
main()
```

```python
def createFreqDict(quote):
    out = wordsFromFullQuotes(quote)
    outDict = {}
    for i in out:
        if i not in outDict.keys():
            outDict[i] = 1
        else:
            outDict[i] += 1
    return outDict
def createPostDict():
    quoteList = listFullOfQuotes()
    postingList = {}
    for item in quoteList:
        #print item
        postingList[item] = createFreqDict(item)
    return postingList
def main():
    print createPostDict()
main()
```

#### d.
    - createReverseDict:
        Creates a Reverse Post Dictionary from the quotes in quote.txt
        Input: 
            None
        Output:
            Reverse Post Dictionary
        

```python
def createReverseDict():
    postDict = createPostDict()
    reverseDict = {}
    for quote in postDict.keys():
        for word in postDict[quote].keys():
            if word not in reverseDict.keys():
                reverseDict[word] = {quote:postDict[quote][word]}
            else:
                reverseDict[word][quote] = postDict[quote][word]
    return reverseDict
def main():
    out = createReverseDict()
    #prints out first 20 values
    for index,key in enumerate(out.keys()):
        print "{"+ key + ":" + str(out[key]) +"}"
        if index == 20:
            break
main()
```

#### f.
    -quoteSearch:
        Takes a word and returns a dictionary where the keys are quotes that contain the word and the values are TD-IDF values of searched word
        Input:

        Output:

```python
def quoteSearch(word):
    #returns a dictionary
    # - keys are full quotes containing that word
    # - values are TF-IDF scores of that word
    reverseDict = createReverseDict()
    #quotes that Contain word
    if word not in reverseDict.keys():
        print "Unable to find word"
        return
    releventQuoteList = reverseDict[word].keys()
    outDict = {}
    for quote in releventQuoteList:
        outDict[quote] = calculateTDIDF(word,quote,reverseDict)
    return outDict
def main():
    print quoteSearch("she")
main()   
```

#### g.
    multiQuoteSearch
        Given a list of words, performs an "AND" search on quotes and sums the TD-IDF frequencies
        Input: 
            wordList: a list of words to search for
        Output:
            Dictionary where keys are the quotes and the values are the sum of the TD-IDF values

```python
def multiQuoteSearch(wordList):
    for i in range(len(wordList)):
        wordList[i] = wordList[i].lower()
    reverseDict =  createReverseDict()
    postDict = createPostDict()
    releventQuoteDict = {}
    for word in wordList:
        if word not in reverseDict.keys():
            print "Invalid Input"
            return ""
        #get all quotes that contain that specific word and their frequencies
        for quote in reverseDict[word].keys():
            if quote not in releventQuoteDict.keys():
                releventQuoteDict[quote] = 1
            else:
                releventQuoteDict[quote] += 1
    #releventQuoteDict now has all quotes that contain the words in the list and their frequencies
    finalQuoteList = []
    for quote in releventQuoteDict.keys():
        #here is the "AND" part of the search
        if releventQuoteDict[quote] == len(wordList):
            finalQuoteList.append(quote)
    outDict = {}
    for quote in finalQuoteList:
        #calculate the sum of the TD-IDF values for all quotes        
        sumTDIDF = 0
        for word in postDict[quote].keys():
            sumTDIDF += calculateTDIDF(word,quote,reverseDict)
        outDict[quote] = sumTDIDF
    return outDict
def main():
    print multiQuoteSearch(["she","he"])
main()
```
